```python
x=np.random.rand(50)
y=np.random.rand(50)
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y,s=20,c="red",alpha=0.5)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Random")
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
color=np.random.rand(50)
plt.scatter(x,y,s=50,c=color,alpha=0.5,marker="*")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Random")
plt.show()
```
